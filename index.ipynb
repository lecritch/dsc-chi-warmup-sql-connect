## Remember how to connect to an sqlite database?

```python
#connect to the db 'parlgov-development' in the data folder
import sqlite3
conn = sqlite3.connect('data/parlgov-development.sqlite')
cur = conn.cursor()
```

### And select data from the table `view_election`?
